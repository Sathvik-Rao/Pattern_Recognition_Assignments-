1. Describe what an identity matrix is. What does multiplying with an identity matrix do?
1. Create an identity matrix called `identity_matrix` that has the same size as `img_average_channel`.

An identity matrix, often denoted as $I$ or $I_n$ (where n represents the size of the square matrix), is a special square matrix in which all the diagonal elements are equal to 1, and all the off-diagonal elements are equal to 0. In mathematical notation, an identity matrix of size n x n is defined as:


$I_n = \begin{bmatrix}
1 & 0 & 0 & \ldots & 0 \\
0 & 1 & 0 & \ldots & 0 \\
0 & 0 & 1 & \ldots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \ldots & 1 \\
\end{bmatrix}$


When we multiply any matrix by an identity matrix of the appropriate size, it acts as a kind of "multiplicative identity" or "neutral element." In other words, the result of the multiplication is the original matrix itself. Mathematically, if $A$ is any matrix of appropriate size, then:


$A \cdot I_n = A$

So, multiplying a matrix by an identity matrix leaves the matrix unchanged. It's similar to multiplying a number by 1; the value remains the same.

```python
# Create an identity matrix with the same size as img_average_channel
identity_matrix = np.eye(img_average_channel.shape[0], img_average_channel.shape[1])

#This above line uses np.eye() to create an identity matrix with the same number of rows and columns as img_average_channel.
```

This code first generates a random row index within the dimensions of `identity_matrix`, and then it prints the contents of that randomly selected row.

Observation: The printed row will contain all zeros except for a single 1 at the position corresponding to the randomly selected row index. This pattern is consistent with the definition of an identity matrix, where all diagonal elements are 1, and all other elements are 0.

We can calculate the determinant of `T` using NumPy's `np.linalg.det` function


```python
# Compute the inverse of T
Tinv = np.linalg.inv(T)

# If `T` is invertible, we can compute its inverse using `np.linalg.inv`
```
